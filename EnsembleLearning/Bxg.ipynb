```python
# %pip install xgboost
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
x, y = load_iris(return_X_y = True)
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3)
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    use_label_encoder=False,   
    eval_metric="mlogloss"     
)

print(xgb_model)
```

```python
xgb_model.fit(x_train,y_train)
```

```python
y_predicted = xgb_model.predict(x_test)
print(y_predicted.shape)
```

```python
print("Accuracy:", accuracy_score(y_test, y_predicted))
print("Classification Report:\n", classification_report(y_test, y_predicted))
```

```python
xgb.plot_importance(xgb_model) #this is the predefined fucntion in the xgb model that can help us in finding out ehich is the ebst feature according to the feature importance 
plt.show()
```
