```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
```

```python
df = pd.read_csv(r"C:\Users\ishaa\OneDrive\Desktop\InternshipGenAI\kaggleDATA\Simple_Stock_Dataset.csv")
print(df.head())
print(df.columns)
print(df.info())
print(df.dtypes)

```

```python
# Step 1: Convert 'Date' column to datetime as usually the date is in the object from there now the date will we converted to the date format 

df['Date'] = pd.to_datetime(df['Date'])
print(df.dtypes)

# Step 2: Plot the closing price
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], color='blue', linestyle='--', marker='o', label='Close Price')
plt.xlabel("Date")
plt.ylabel("Price (â‚¹)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#barChart using SubPlots 
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Step 2: Plot closing price
axs[0].plot(df['Date'], df['Close'], color='blue', marker='o', label='Close Price')
axs[0].set_title("Stock Closing Price")
axs[0].set_ylabel("Price ($)")
axs[0].legend()
axs[0].grid(True)

# Step 3: Plot volume
axs[1].bar(df['Date'], df['Volume'], color='Black')
axs[1].set_title("Volume Traded")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Volume")
axs[1].grid(True)

# Step 4: Adjust layout
plt.tight_layout()
plt.show()
```
